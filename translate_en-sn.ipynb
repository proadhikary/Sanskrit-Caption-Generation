```python
import time
import pandas as pd
from deep_translator import GoogleTranslator

data = pd.read_csv('captions8K.txt')

for index, row in data.iterrows():
    original_text = row['caption']
    translated_text = GoogleTranslator(source='auto', target='sanskrit').translate(text=original_text)
    print(index)
    with open("sn8K.txt", 'a') as f:
      f.write(translated_text + '\n')
```
